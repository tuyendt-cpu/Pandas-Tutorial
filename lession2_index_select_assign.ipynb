<h3 style="color: red; font-family: Monaco">Indexing in pandas </h3>

* **Index - based selection** <br/>
Selecting data based on its numberical postion in the data

```python
reviews.iloc[0]
```

```python
#Row first, Column second
reviews.iloc[:3,:3 ]
```

* **Label-based selection** <br/>
It use index value

```python
reviews.loc[0, 'country']
```

```python
reviews.loc[:5, ['taster_name', 'taster_twitter_handle', 'points']]
```

```python
reviews.describe()
```

```python
reviews.set_index('title')
```

```python
#return a Series base on country column
reviews.country == 'Italy' 
```

```python
#the above result can be used inside of loc to select the correct
reviews.loc[reviews.country == 'Italy']
```

```python
#we can use & to bring 2 question together
reviews.loc[(reviews.country =='Italy') & (reviews.points >=90)]
```

```python
#we can use | as or operator
reviews.loc[(reviews.country == 'Italy')|(reviews.points >=90)]
```

```python
# `isin` is built-in conditional in pandas. It let you select
# data whose value is in a list of values 
reviews.loc[reviews.country.isin(['Italy', 'France'])]
```

```python
#The second is isnull. These method let you highlight values which
#are empty. 
#Example: filter wines lacking a price tag
reviews.loc[reviews.price.notnull()]
```
